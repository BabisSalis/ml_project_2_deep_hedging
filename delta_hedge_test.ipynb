```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
from src.env import StockTradingEnv
from src.network import MLP
```

```python
env = StockTradingEnv(True)

actor = MLP(3, 32, 1, "Sigmoid")
#actor.load_state_dict(torch.load(f"model/v5/actor_weight.pt"))
```

```python
state = env.reset()

tot_reward = 0
while True:
    action = actor(torch.Tensor(state)).detach().item() * 100
    action = np.round(np.clip(action, 0, 100), 2)
    
    next_state, reward, done = env.step(action)
    
    print(state, action, next_state, reward, done)
    tot_reward += reward 
    state = next_state
    if done:
        break
    
```
